```python
import numpy as np
import matplotlib.pyplot as plt
from cv2 import imshow, imread


def show(im1, im2):
    plt.figure(figsize=(10, 20))
    plt.subplot(1, 2, 1)
    plt.axis('off')
    plt.tight_layout()
    plt.imshow(im1)

    plt.subplot(1, 2, 2)
    plt.axis('off')
    plt.tight_layout()
    plt.imshow(im2)
    plt.show()

```

```python
arr = []
with open('pictures/scores.txt') as f:
    for line in f:
       arr.append([float(x) for x in line.replace('\n', '').split(',')])
arr = np.array(arr)
```

```python
for j, i in enumerate((5, 10, 25, 50, 100)):
    print(f'number of keywords is {i}')
    print(f'test accuracy {arr[j, 1]}', '\n')
    im1 = imread(f'pictures/top{str(i)}_1.png')
    im2 = imread(f'pictures/top{str(i)}_2.png')
    show(im1, im2)
    print('')
```

```python
plt.plot(arr[:, 0], arr[:, 1])
plt.xlabel("number of keywords")
plt.ylabel("test accuracy")
plt.grid('on')
plt.ylim(0, 75)
None
```
